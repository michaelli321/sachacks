# Age

```python
plt.hist(np.delete(np.nan_to_num(np.array(data['Age'])), np.where(np.nan_to_num(np.array(data['Age'])) == 0)))
```

```python
male_ages = {}
female_ages = {}
male_counted = {}
female_counted = {}

for i, row in data.iterrows():
    if row['Sex'] == 'M':
        if row['Year'] not in male_ages:
            male_ages[row['Year']] = [row['Age']]
            male_counted[row['Year']] = {row['Name']}
        else:
            if row['Name'] in male_counted[row['Year']]:
                continue
            else:
                male_ages[row['Year']].append(row['Age'])
                male_counted[row['Year']].add(row['Name'])
    else:
        if row['Year'] not in female_ages:
            female_ages[row['Year']] = [row['Age']]
            female_counted[row['Year']] = {row['Name']}
        else:
            if row['Name'] in female_counted[row['Year']]:
                continue
            else:
                female_ages[row['Year']].append(row['Age'])
                female_counted[row['Year']].add(row['Name'])
```

```python
male_ages_missing = {}
female_ages_missing = {}

for key in male_ages:
    male_ages[key] = np.nan_to_num(male_ages[key])
for key in female_ages:
    female_ages[key] = np.nan_to_num(female_ages[key])
    
for key in male_ages:
    male_ages_missing[key] = list(male_ages[key]).count(0)
    male_ages[key] = male_ages[key][male_ages[key] != 0]
for key in female_ages:
    female_ages_missing[key] = list(female_ages[key]).count(0)
    female_ages[key] = female_ages[key][female_ages[key] != 0]
```

```python
male_avgs = []
female_avgs = []

for key in male_ages:
    male_avgs.append(np.mean(male_ages[key]))
for key in female_ages:
    female_avgs.append(np.mean(female_ages[key]))
```

```python
plt.scatter(list(male_ages.keys()), male_avgs)

for year in male_ages_missing:
    print(str(year) + " missing " + str(male_ages_missing[year]) + " values ")
```

```python
plt.scatter(list(female_ages.keys()), female_avgs)

for year in female_ages_missing:
    print(str(year) + " missing " + str(female_ages_missing[year]) + " values ")
```

<b>Interesting, theres a general increase in age in recent history. The pre 1960 phase can be ignored as there are so many missing values</b>
