# Age

```python
plt.hist(np.delete(np.nan_to_num(np.array(data['Age'])), np.where(np.nan_to_num(np.array(data['Age'])) == 0)))
```

```python
male_ages = {}
female_ages = {}
male_counted = {}
female_counted = {}

for i, row in data.iterrows():
    if row['Sex'] == 'M':
        if row['Year'] not in male_ages:
            male_ages[row['Year']] = [row['Age']]
            male_counted[row['Year']] = {row['Name']}
        else:
            if row['Name'] in male_counted[row['Year']]:
                continue
            else:
                male_ages[row['Year']].append(row['Age'])
                male_counted[row['Year']].add(row['Name'])
    else:
        if row['Year'] not in female_ages:
            female_ages[row['Year']] = [row['Age']]
            female_counted[row['Year']] = {row['Name']}
        else:
            if row['Name'] in female_counted[row['Year']]:
                continue
            else:
                female_ages[row['Year']].append(row['Age'])
                female_counted[row['Year']].add(row['Name'])
```

```python
male_ages_missing = {}
female_ages_missing = {}

for key in male_ages:
    male_ages[key] = np.nan_to_num(male_ages[key])
for key in female_ages:
    female_ages[key] = np.nan_to_num(female_ages[key])
    
for key in male_ages:
    male_ages_missing[key] = list(male_ages[key]).count(0)
    male_ages[key] = male_ages[key][male_ages[key] != 0]
for key in female_ages:
    female_ages_missing[key] = list(female_ages[key]).count(0)
    female_ages[key] = female_ages[key][female_ages[key] != 0]
```

```python
male_avgs = []
female_avgs = []

for key in male_ages:
    male_avgs.append(np.mean(male_ages[key]))
for key in female_ages:
    female_avgs.append(np.mean(female_ages[key]))
```

```python
plt.scatter(list(male_ages.keys()), male_avgs)

for year in male_ages_missing:
    print(str(year) + " missing " + str(male_ages_missing[year]) + " values ")
```

```python
plt.scatter(list(female_ages.keys()), female_avgs)

for year in female_ages_missing:
    print(str(year) + " missing " + str(female_ages_missing[year]) + " values ")
```

<b>Interesting, theres a general increase in age in recent history. The pre 1960 phase can be ignored as there are so many missing values</b>

# Country Medals

```python
country_medals = {}

for i, row in data.iterrows():
    if row['Team'] not in country_medals:
        country_medals[row['Team']] = {
            'Gold': 0,
            'Silver': 0,
            'Bronze': 0
        }
        
    if not pd.isnull(row['Medal']):
        country_medals[row['Team']][row['Medal']] += 1
```

```python
countries = []
bronze_counts = []
silver_counts = []
gold_counts = []

for key in country_medals:
    countries.append(key)
    bronze_counts.append(country_medals[key]['Bronze'])
    silver_counts.append(country_medals[key]['Silver'])
    gold_counts.append(country_medals[key]['Gold'])
    
max_is = np.argsort(bronze_counts)[-5:]
max_bronze = []
max_countries = []
for i in max_is:
    max_bronze.append(bronze_counts[i])
    max_countries.append(countries[i])
plt.bar(max_countries, max_bronze)
plt.show()
```

```python
max_is = np.argsort(silver_counts)[-5:]
max_silver = []
max_countries = []
for i in max_is:
    max_silver.append(silver_counts[i])
    max_countries.append(countries[i])
plt.bar(max_countries, max_silver)
plt.show()
```

```python
max_is = np.argsort(gold_counts)[-5:]
max_gold = []
max_countries = []
for i in max_is:
    max_gold.append(gold_counts[i])
    max_countries.append(countries[i])
plt.bar(max_countries, max_gold)
plt.show()
```

```python
tot_counts = []
for i in range(len(bronze_counts)):
    tot = bronze_counts[i] + silver_counts[i] + gold_counts[i]
    tot_counts.append(tot)
    
max_is = np.argsort(tot_counts)[-5:]
max_tot = []
max_countries = []
for i in max_is:
    max_tot.append(tot_counts[i])
    max_countries.append(countries[i])
plt.bar(max_countries, max_tot)
plt.show()
```

```python
medals_over_years = {}
countries_of_interest = {'United States', 'Soviet Union', 'Germany', 'Great Britain', 'France'}
for i, row in data.iterrows():
    if row['Year'] not in medals_over_years:
        medals_over_years[row['Year']] = {
            'United States': 0,
            'Soviet Union': 0,
            'Germany': 0,
            'Great Britain': 0,
            'France': 0
        }
        
    if not pd.isnull(row['Medal']):
        if row['Team'] in countries_of_interest:
            medals_over_years[row['Year']][row['Team']] += 1
```

```python
years = []
france_medals = []
germany_medals = []
soviet_medals = []
united_medals = []
britain_medals = []

for year in medals_over_years:
    years.append(year)
    france_medals.append(medals_over_years[year]['France'])
    germany_medals.append(medals_over_years[year]['Germany'])
    soviet_medals.append(medals_over_years[year]['Soviet Union'])
    united_medals.append(medals_over_years[year]['United States'])
    britain_medals.append(medals_over_years[year]['Great Britain'])
```

```python
order_years = []
order_france = []
order_germany = []
order_soviet = []
order_united = []
order_britain = []
ordered_is = np.argsort(years)

for i in ordered_is:
    order_years.append(years[i])
    order_france.append(france_medals[i])
    order_germany.append(germany_medals[i])
    order_soviet.append(soviet_medals[i])
    order_united.append(united_medals[i])
    order_britain.append(britain_medals[i])
```

```python
plt.plot(order_years, order_france, color='blue')
plt.plot(order_years, order_germany, color='green')
plt.plot(order_years, order_soviet, color='red')
plt.plot(order_years, order_united, color='purple')
plt.plot(order_years, order_britain, color='yellow')
```

```python
france_medals = []
germany_medals = []
soviet_medals = []
united_medals = []
britain_medals = []

for i in range(len(order_years)):
    if i == 0:
        france_medals.append(order_france[i])
        germany_medals.append(order_germany[i])
        soviet_medals.append(order_soviet[i])
        united_medals.append(order_united[i])
        britain_medals.append(order_britain[i])
    else:
        france_medals.append(order_france[i] + france_medals[-1])
        germany_medals.append(order_germany[i] + germany_medals[-1])
        soviet_medals.append(order_soviet[i] + soviet_medals[-1])
        united_medals.append(order_united[i] + united_medals[-1])
        britain_medals.append(order_britain[i] + britain_medals[-1])
```

```python
plt.plot(order_years, france_medals, color='blue')
plt.plot(order_years, germany_medals, color='green')
plt.plot(order_years, soviet_medals, color='red')
plt.plot(order_years, united_medals, color='purple')
plt.plot(order_years, britain_medals, color='yellow')
```
